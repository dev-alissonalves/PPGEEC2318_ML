#### Artigo: Sobre o impacto social dos modelos de fundações abertas


* **Aspectos positivos**
_____________________________________

Compartilhar modelos em Aprendizado de Máquina traz uma série de benefícios importantes tanto para a comunidade científica quanto para a sociedade em geral. Uma das grandes vantagens é estimular a colaboração na pesquisa entre universidades, institutos de ensino e escolas técnicas, permitindo que cientistas ao redor do mundo contribuam e se beneficiem dos avanços uns dos outros. Isso resulta em uma aceleração da inovação, pois as diferentes contribuições se somam para resolver desafios mais complexos de maneira mais eficaz e rápida. Além disso, o compartilhamento de modelos amplia o acesso a tecnologias de ponta, possibilitando que profissionais de diferentes níveis de experiência e recursos (*na minha opinião, principalmente computacionais*) utilizem modelos de alta qualidade em seus projetos. Essa democratização do acesso não só enriquece a pesquisa com diversas perspectivas, mas também favorece o desenvolvimento de soluções mais inclusivas e socialmente relevantes. Neste último ponto, destaco o trecho da entrevista do Enio Morais, CIO na Semantrix, onde ele afirma que: *"A inteligência artificial tem possibilitado a democratização de serviços de várias maneiras. Dentre elas, é necessário destacar o acesso igualitário à informação."*

* **Aspectos negativos**
_____________________________________

O compartilhamento de modelos, especialmente quando se trata de modelos de base aberta, pode apresentar algumas desvantagens e problemas significativos. Uma das principais preocupações é o risco de uso indevido associado a estes modelos. Por exemplo, podem ser utilizados para atividades maliciosas, como ataques cibernéticos, difusão de desinformação, clonagem de vozes para fraudes, ou mesmo para abuso infantil e imagens íntimas não consensuais. Neste último exemplo, podemos citar o acontecimento no Rio de Janeiro em novembro do ano passado onde alunos de uma escola fizeram uso da inteligência artificial para fins ilegais.

 [ARTIGO NA ÍNTEGRA](https://www.cnnbrasil.com.br/nacional/alunos-de-colegio-tradicional-do-rio-usam-ia-para-criar-imagens-intimas-de-meninas-policia-investiga/)



Além disso, existem questões éticas associadas aos compartilhamento de modelos, especialmente quando não há controle suficiente sobre a forma como serão utilizados. Isto pode incluir preocupações com a privacidade, no que se diz respeito a forma como nossos dados estão sendo utilizados por grandes companhias, preconceitos algorítmicos e justiça social, especialmente quando os modelos são utilizados de forma discriminatória ou injusta.
